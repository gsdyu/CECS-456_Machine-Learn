```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
beta_0 = 5; beta_1 = 1
```

```python
#size of 100 in a 10x10 matrix
X = np.random.normal(0, 1, 100)
```

```python
#size of 100 in a 10x10 matrix
noise = np.random.normal(0, 1, 100)
```

```python
Y = beta_0 + beta_1*X+noise
```

```python
'''X design matrix of format: [[1, x1]
                               [1, x2]
                               ...
                               [1, xN]]
'''
X_design = np.column_stack([np.ones(100),X])
```

```python
B_hat = np.linalg.inv(X_design.transpose()@X_design)@(X_design.transpose()@Y)
```

```python
#Extra Credit/Visualize

#using predicted B_hat
Y_pred = B_hat[0] + B_hat[1]*X

Y_without_noise = beta_0 + beta_1*X

#set up grid
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')

#raw data
plt.scatter(X, Y, color="blue")

#line of regression
plt.plot(X, Y_without_noise, color="green", label="line of regression")
plt.plot(X, Y_pred, color="red", label="line of regression (prediction)")

#set up table
plt.legend()

```
